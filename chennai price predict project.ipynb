```python
df1=pd.read_csv("C:\\Users\prath\prathik ranka\python\simple programs\chennai_rate_prediction\chennai property dataset.csv")
df1.head()
```

```python
df1.shape
```

```python
df2=df1.drop(columns=['status','builder','age'])
df2.head(10)
```

```python
df2.isnull().sum()
```

```python
df2.bathroom=df2.bathroom.fillna(df2.bathroom.median())
```

```python
df2
```

```python
len(df2.location.unique())
```

```python
df2['bhk'].unique()
```

```python
dummies=pd.get_dummies(df2.location)
dummies.head(25)
```

```python
df3=pd.concat([df2,dummies],axis='columns')
df3
```

```python
x=df3.drop(columns=['price','location'])
x.iloc[20]
```

```python
y=df3.price
y.iloc[20]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
Model=LinearRegression()
Model.fit(x_train,y_train)
Model.score(x_train,y_train)
```

```python
def predict_price(location,area,bhk,bath):
    loc_index=np.where(x.columns==location)[0][0]

    z=np.zeros(len(x.columns))
    z[0]=area
    z[1]=bhk
    z[2]=bath
    if loc_index>=0:
        z[loc_index]=1
    
    return Model.predict([z])[0]
```

```python
predict_price('Mogappair',2225,3,2)
```

```python
predict_price('Veppampattu',1200,3,1)
```
